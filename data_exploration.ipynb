```python
import pandas as pd
import requests


INTEREST_RATES_API = "https://www.mas.gov.sg/api/v1/bondsandbills/m/savingbondsinterest"
SSB_INFO_API = "https://www.mas.gov.sg/api/v1/bondsandbills/m/listsavingbonds"
```

```python
# Getting the total number of SSBs in history
resp = requests.get(f"{INTEREST_RATES_API}?rows=1")
if resp.status_code == 200:
    num_ssb = resp.json()['result']['total']
else:
    raise AssertionError(f'Failed API request')
```

```python
interest_rates_df = requests.get(f"{INTEREST_RATES_API}?rows={num_ssb}").json()['result']['records']
interest_rates_df = pd.DataFrame(interest_rates_df)
ssb_info_df = requests.get(f"{SSB_INFO_API}?rows={num_ssb}").json()['result']['records']
ssb_info_df = pd.DataFrame(ssb_info_df)
```
